### Scraping Bitcoin Tweets

```python
# Search Query (filter out the retweets)
search_query = "#btc -filter:retweets"

# Get tweets from the API
tweets = tw.Cursor(api.search,
              q=search_query,
              lang="en",
              since="2020-09-16").items(500)

# store the API responses in a list
tweets_copy = []
for tweet in tweets:
    tweets_copy.append(tweet)
    
print("Total Tweets fetched:", len(tweets_copy))
```

```python
# intialize the dataframe
tweets_btc = pd.DataFrame()

# populate the dataframe
for tweet in tweets_copy:
    hashtags = []
    try:
        for hashtag in tweet.entities["hashtags"]:
            hashtags.append(hashtag["text"])
        text = api.get_status(id=tweet.id, tweet_mode='extended').full_text
    except:
        pass
    tweets_btc = tweets_btc.append(pd.DataFrame({"tweet_id": tweet.id,
                                               "name": tweet.user.name,
                                               "screen_name": tweet.user.screen_name,
                                               "retweet_count": tweet.retweet_count,
                                               "text": text,
                                                 "date": tweet.created_at,
                                               "created_at": tweet.created_at,
                                               "favourite_count": tweet.favorite_count,
                                                 "hashtags": tweet.entities["hashtags"],
                                                 "status_count": tweet.user.statuses_count,
                                                 "followers_count": tweet.user.followers_count,
                                                 "location": tweet.place,
                                                 "source_device": tweet.source,}))
    tweets_btc = tweets_btc.reset_index(drop=True)


# show the dataframe
tweets_btc.head()
```

```python
# clean up tweets - keep text, retweet_count, favourite_count, and followers_count
tweets_btc = tweets_btc[['tweet_id', 'date', 'text', 'retweet_count', 'favourite_count', 'followers_count']]
tweets_btc['crypto_name'] = 'BTC'
tweets_btc.head()
```

```python
tweets_btc.info()
```

```python
## Export BTC tweets
tweets_btc.to_csv('data/tweets_btc_5.csv')
```

### Load all the crypto tweets and combine them

```python
btc_1 = pd.read_csv('data/tweets_btc_1.csv')
btc_2 = pd.read_csv('data/tweets_btc_1.csv')
btc_3 = pd.read_csv('data/tweets_btc_1.csv')
btc_4 = pd.read_csv('data/tweets_btc_1.csv')
btc_5 = pd.read_csv('data/tweets_btc_1.csv')
```

```python
eth_1 = pd.read_csv('data/tweets_eth_1.csv')
eth_2 = pd.read_csv('data/tweets_eth_1.csv')
eth_3 = pd.read_csv('data/tweets_eth_1.csv')
eth_4 = pd.read_csv('data/tweets_eth_1.csv')
eth_5 = pd.read_csv('data/tweets_eth_1.csv')
```

```python
ada_1 = pd.read_csv('data/tweets_ada_1.csv')
ada_2 = pd.read_csv('data/tweets_ada_1.csv')
ada_3 = pd.read_csv('data/tweets_ada_1.csv')
ada_4 = pd.read_csv('data/tweets_ada_1.csv')
ada_5 = pd.read_csv('data/tweets_ada_1.csv')
```

```python
dog_1 = pd.read_csv('data/tweets_dog_1.csv')
dog_2 = pd.read_csv('data/tweets_dog_1.csv')
dog_3 = pd.read_csv('data/tweets_dog_1.csv')
dog_4 = pd.read_csv('data/tweets_dog_1.csv')
dog_5 = pd.read_csv('data/tweets_dog_1.csv')
```

### Combines all the tweets and drop duplicates

```python
# combined all BTC
btc_df = pd.concat([btc_1, btc_2, btc_3, btc_4, btc_5], ignore_index=True)

# drop dubplicates
btc_df.drop_duplicates(subset=['tweet_id'], inplace=True, keep='last', ignore_index=True)

# covert date column to datetime
btc_df['date'] = pd.to_datetime(btc_df['date'])

# drop Unnamed: 0 column
btc_df.drop(columns='Unnamed: 0', inplace=True)

# add an empty column called 'manual_sentiment'
#combined_df['manual_sentiment'] = ''

# export btc_df
btc_df.to_csv('data/btc_df_complete.csv')

btc_df.info()
```

```python
# combined all ETH
eth_df = pd.concat([eth_1, eth_2, eth_3, eth_4, eth_5], ignore_index=True)

# drop dubplicates
eth_df.drop_duplicates(subset=['tweet_id'], inplace=True, keep='last', ignore_index=True)

# covert date column to datetime
eth_df['date'] = pd.to_datetime(eth_df['date'])

# drop Unnamed: 0 column
eth_df.drop(columns='Unnamed: 0', inplace=True)

# export btc_df
eth_df.to_csv('data/eth_df_complete.csv')

eth_df.info()
```

```python
# combined all ADA
ada_df = pd.concat([ada_1, ada_2, ada_3, ada_4, ada_5], ignore_index=True)

# drop dubplicates
ada_df.drop_duplicates(subset=['tweet_id'], inplace=True, keep='last', ignore_index=True)

# drop nulls
ada_df.dropna(axis = 0, how ='any', inplace=True)

# covert date column to datetime
ada_df['date'] = pd.to_datetime(ada_df['date'])

# convert retweet_count to float64
ada_df['retweet_count'] = pd.to_numeric(ada_df['retweet_count'])

# drop Unnamed: 0 column
ada_df.drop(columns='Unnamed: 0', inplace=True)

# export btc_df
ada_df.to_csv('data/ada_df_complete.csv')

ada_df.info()
```

```python
# combined all DOG
dog_df = pd.concat([dog_1, dog_2, dog_3, dog_4, dog_5], ignore_index=True)

# drop dubplicates
dog_df.drop_duplicates(subset=['tweet_id'], inplace=True, keep='last', ignore_index=True)

# drop nulls
dog_df.dropna(axis = 0, how ='any', inplace=True)

# covert date column to datetime
dog_df['date'] = pd.to_datetime(dog_df['date'])

# convert retweet_count to float64
dog_df['retweet_count'] = pd.to_numeric(dog_df['retweet_count'])

# drop Unnamed: 0 column
dog_df.drop(columns='Unnamed: 0', inplace=True)

# export btc_df
dog_df.to_csv('data/dog_df_complete.csv')

dog_df.info()
```

```python
# combined all BTC + ETH + ADA + DOG
combined_all_df = pd.concat([btc_df, eth_df, ada_df, dog_df], ignore_index=True)

# export to csv
combined_all_df.to_csv('data/all_crypto_tweets.csv', index=False)

combined_all_df.head()
combined_all_df.info()
```
