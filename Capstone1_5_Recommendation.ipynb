### Recommend Buy/Sell

Below we show moving average for 2 days, 5 days, 7 days, and 14 days by using .diff() to calculates the difference in price between a value at time T=2, T=5, T-7, and T=14.

- If the difference is positive, it will return 1.
- If the difference is negative, it will return -1.
- If the difference is neither, it will return 0.

```python
# BTC
btc = df_btc[['Date', 'Marketcap', 'Close']]
btc['target_2days'] = btc['Close'].diff(2).apply(lambda x:-1 if x<0 else (1 if x>0 else 0))
btc['target_5days'] = btc['Close'].diff(5).apply(lambda x:-1 if x<0 else (1 if x>0 else 0))
btc['target_7days'] = btc['Close'].diff(7).apply(lambda x:-1 if x<0 else (1 if x>0 else 0))
btc['target_14days'] = btc['Close'].diff(14).apply(lambda x:-1 if x<0 else (1 if x>0 else 0))
btc.head(20)
```
